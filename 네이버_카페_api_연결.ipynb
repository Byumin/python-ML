```python
import os
import sys
import urllib.request
import pandas as pd
import json
import re # 정규 및 표현식 활용

client_id = "OtQDai4tGvlqkAoSQa3e"
client_secret = "_XIdXBNyul"

query = urllib.parse.quote(input("검색어를 입력하세요 : "))
idx =0
display=100
start =1
end =1000

web_df=pd.DataFrame(columns=("title", "link", "description", "cafename", "cafeurl"))

for start_index in range(start, end, display):
  url = "https://openapi.naver.com/v1/search/cafearticle?query=" + query + "&display=" + str(display) + "&start=" + str(start_index) # JSON 결과
  # url = "https://openapi.naver.com/v1/search/cafearticle.xml?query=" + encText # XML 결과

  request = urllib.request.Request(url)
  request.add_header("X-Naver-Client-Id",client_id)
  request.add_header("X-Naver-Client-Secret",client_secret)
  response = urllib.request.urlopen(request) # http에 요청을 하고 그 결과를 response에 저장
  rescode = response.getcode() # response에 저장한 결과(http 응답)에서 상태코드를 가지고와 rescode에 저장
  if(rescode==200): # rescode의 값이 200인 경우 http에서 성공을 의미
      response_body = response.read() #http 응답의 본문을 읽어서 response_body에 저장
      remove_tag = re.compile("<.*?>")
      response_body_str = response_body.decode('utf-8') # response_body utf-8로 디코드해서 response_body_str 저장
      response_body_remove = re.sub(remove_tag, "", response_body_str) # response_body_str에서 remove_tag 해당되는 정규 표현식 대체
      response_body_dict = json.loads(response_body_remove)
      # 문제 : display 100 설정으로 df 만들었을 때 마지막 901부터 1000까지의 값만 데이터 프레임화 되고 끝나게 됨. -> 해당 display된것을 df화 해 web_df(메타 데이터 프레임)에 병합으로 해결
      df = pd.DataFrame(response_body_dict['items'])
      web_df = pd.concat([web_df, df])
# 보통 json.loads해서 딕셔너리를 데이터 프레임화 해서 정규표현식으로 for문으로 하나씩 돌려서 제거하지만 이렇게 되면 소요시간이 발생
  else:
      print("Error Code:" + rescode)
```
